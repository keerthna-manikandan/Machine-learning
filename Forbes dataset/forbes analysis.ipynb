```python
#import lib

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("Billionaire.csv")
data
```

```python
data.isnull().sum()
```

```python
data.dropna()
```

```python
data["NetWorth"]= data["NetWorth"].str.strip('$')
data["NetWorth"]= data["NetWorth"].str.strip('B')
data["NetWorth"]= data["NetWorth"].astype(float)
```

```python
data["NetWorth"]
```

```python
df=data.sort_values(by=["NetWorth"],ascending=False).head(15)
plt.figure(figsize=(20,10))
plt.bar(df['Name'],df['NetWorth'])
plt.show()
```

```python
a=data["Source"].value_counts().head(5)
index=a.index
print(index)
sources = a.values
plt.figure(figsize=(5,5))
plt.pie(sources,labels=index)
center=plt.Circle((0,0),0.5,color='white')
fig=plt.gcf()
fig.gca().add_artist(center)
plt.rc('font',size=12)
plt.title("top 5 domains with the most number of billionaires",fontsize=30)
plt.show()
```

```python

col = ["tomato","chocolate","purple","skyblue","yellow"]
a=data["Industry"].value_counts().head(5)
index=a.index
print(index)
sources=a.values
print(sources)
plt.figure(figsize=(5,5))
plt.pie(sources,labels=index,colors=col)
center=plt.Circle((0,0),0.5,color='white')
fig=plt.gcf()
fig.gca().add_artist(center)
plt.rc('font',size=12)
plt.title("top 5 industries with the most number of billionaires",fontsize=30)
plt.show()
```
